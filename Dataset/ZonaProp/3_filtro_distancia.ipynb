```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium as f
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut


from sklearn.cluster import k_means
```

```python
df = pd.read_csv('/content/Mercado-inmobilirario-cba/Dataset/ZonaProp/data_final.csv').drop('Unnamed: 0', axis=1)
df.head()
```

```python
cordoba_coords = (-31.4135, -64.18105)

# Distancia de referencia (en kilómetros)
distancia_referencia_km = 50

# Filtrar los puntos que están a una distancia mayor que la distancia de referencia
df['Distancia_a_Cordoba_km'] = df.apply(lambda row: geodesic(cordoba_coords, (row['Latitud'], row['Longitud'])).kilometers, axis=1)
df_filtrado = df[df['Distancia_a_Cordoba_km'] <= distancia_referencia_km]
```

```python
df_filtrado
```

```python
out = df[df['Distancia_a_Cordoba_km'] >= distancia_referencia_km]
out
```

```python
def obtener_coordenadas(ubicacion):
    geolocator = Nominatim(user_agent="myGeocoder")
    try:
        location = geolocator.geocode(ubicacion, timeout=5)
        return (location.latitude, location.longitude) if location else (None, None)
    except GeocoderTimedOut:
        return obtener_coordenadas(ubicacion)
```
